```python
(train_data,train_labels),(test_data,test_labels) = keras.datasets.boston_housing.load_data()
```

```python
import pandas as pd
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
                'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
df = pd.DataFrame(train_data,columns=column_names)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df = sc.fit_transform(df)
```

```python
test = sc.transform(test_data)
```

```python
def build_model():
    model = keras.models.Sequential([
        keras.layers.Dense(64,activation=tf.nn.relu,input_shape=[13]),
        keras.layers.Dense(64, activation=tf.nn.relu),
        keras.layers.Dense(1)
    ])
    optimizer = tf.train.RMSPropOptimizer(0.001)
    model.compile(loss='mse',optimizer=optimizer,metrics=['mae'])
    return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
EPOCHS=500
tb = keras.callbacks.TensorBoard(log_dir='Graph/Boston_housing/',histogram_freq=0
                           ,write_graph=True,write_images=True)
history = model.fit(df, train_labels, epochs=EPOCHS,
                    validation_split=0.2, verbose=1,
                    callbacks=[tb])
```

```python
import matplotlib.pyplot as plt
test_data = sc.transform(test_data)
test_predictions = model.predict(test_data).flatten()

plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [1000$]')
plt.ylabel('Predictions [1000$]')
plt.axis('equal')
plt.xlim(plt.xlim())
plt.ylim(plt.ylim())
_ = plt.plot([-100, 100], [-100, 100])
```
